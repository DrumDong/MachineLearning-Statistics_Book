# Decision Tree 트리 구성 규칙: ID3
-  트리의 핵심은 변별력이 좋은 질문을 위에서부터 하나하나 세팅하는 것
-  정보 이론의 정보량과 엔트로피를 활용
 - 엔트로피는 정보량의 기댓값(평균)을 나타낸다. 발생한 사건들의 정보량을 모두 구해서 (가중)평균낸 것
 - 정보량: 어떤 사건이 가지고 있는 정보의 양인데, 드물게 발생하는 일일수록 정보량이 크다.
 

ID3 알고리즘 정지조건
- 해당 노드에 속하는 레코드들이 모두 같은 클래스를 같는다.
- 상위 노드에서 이미 모든 속성들을 다 사용하였다. 더 이상 사용할 속성이 없다.

ID3의 특징
- 범주형 속성에 대해서만 사용이 가능하다.
  (엔트로피 기반 척도 수식이 범주형 속성에 대해서만 적용이 가능하기 때문에 그렇다.) 
- 상위 노드에서 사용된 속성은 다시 사용하지 않는다.

ID3공식.PNG

- S: 이미 발생한 사건의 모음
- c: 사건의 갯수

p(xi)가 분모에 있기 때문에 이를 -1승으로 올리면
- Σ - p(i)*log(p(i)) 로 변환이 가능함.

엔트로피가 크다는 것은 평균 정보량이 크다는 것인데, 대개 사건들이 일어날 확률이 비슷한 경우에 엔트로피가 크다.

앤트로피 변화.PNG

0.5의 확률이 가장 큰 앤트로피를 보여준다. 결국, 앤트로피가 가장 작은 것을 상위 의사결정 노드에 위치시켜야한다. 이것이 바로 ID3의 핵심 포인트이다.

이를 위해 ID3에서는 정보획득이라는 개념을 들여놓았다. 정보획득량은 다음과 같이 계산한다. 
- IG(S,A) = E(S)−E(S|A)

## ID3 알고리즘 설명 예제

다운로드.png

E(경기|날씨) = 맑음 확률 X E(경기|맑음) + 흐림 확률 X E(경기|흐림) + 비 확률 X E(경기|흐림)

캡처.PNG

E(경기) = 0.940 이므로, 0.940- 0.694 = 0.246

즉, IG(경기, 날씨)  = E(경기) - E(경기|날씨) = 0.246

날씨노드가 뿌리 노드(루트 노드)가 된다.

루트노드.PNG

이후 맑음, 흐림, 비로 인덱싱된 상태에서 다시 E(경기)를 구한 후에, 다른 변수로 엔트로피를 구한다.
