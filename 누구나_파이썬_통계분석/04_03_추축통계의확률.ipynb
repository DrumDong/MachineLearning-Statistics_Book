```python
df = pd.read_csv('data/ch4_scores400.csv') # 학생 예시 Score
scores  = np.array(df['score'])
```

계급폭을 1점으로 한 히스토그램

```python
fig = plt.figure(figsize=(10,6))
ax = fig.add_subplot(111)
ax.hist(scores,bins=100,range=(0,100),density=True)
ax.set_xlim(20,100)
ax.set_ylim(0,0.042)
ax.set_xlabel('score')
ax.set_ylabel('relative frequency')
plt.show()
```

4%의 확률로 69점의 표본 데이터를 얻을 수 있음.
- 상대도수(relative frequency)는 해당 점수를 얻는 확률에 대응함. -> 결국, 이 히스토그램은 모집단의 확률분포로 간주 가능

무작위추출은 이와같이 확률분포를 따르는 확률변수의 시행임

```python
np.random.choice(scores)
```

주사위의 상대도수가 시행 횟수를 늘리면 실제의 확률분포에 가까워지는 것처럼,무작위추출에서 표본의 크기가 커지면, 표본 데이터 상대도수는 실제 확률분포에 가까워진다.

```python
sample = np.random.choice(scores,10000)

fig = plt.figure(figsize=(10,6))
ax = fig.add_subplot(111)
ax.hist(sample,bins=100,range=(0,100),density=True)
ax.set_xlim(20,100)
ax.set_ylim(0,0.042)
ax.set_xlabel('score')
ax.set_ylabel('relative frequency')
plt.show()
```

히스토그램이 실제 점수 분포에 매우 가까운 형태임. -> 표본이 커지면, 실제 분포에 수렴

마지막으로 표본평균에 대해 고민해 보자면, 표본 하나하나가 확률변수 이므로, ***표본들의 평균으로 계산되는 표본평균도 역시 확률변수 이다.***

```python
sample_means = [np.random.choice(scores,20).mean() for _ in range(10000)]

fig = plt.figure(figsize=(10,6))
ax = fig.add_subplot(111)
ax.hist(sample_means,bins=100,range=(0,100),density=True)
#모평균을 세로선으로 표시
ax.vlines(np.mean(scores),0,1,'gray')
ax.set_xlim(50,90)
ax.set_ylim(0,0.13)
ax.set_xlabel('score')
ax.set_ylabel('relative frequency')
plt.show()
```

표본평균은 산포도 그 자체이므로, 모평균을 중심으로 분포하는 것을 알 수 있음. 이것이 무작위추출에 의한 표본평균으로 모평균을 추측할 수 있는 근거이다.
