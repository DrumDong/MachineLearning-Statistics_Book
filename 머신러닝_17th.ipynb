---

연관규칙은 장바구니 분석에 사용되는 기계학습의 매우 중요한 개념중 하나이다. 가게에서는 모든 야채가 같은 통로에 놓여지고, 모든 유제품이 함께 놓여지며, 화장품은 그러한 그룹의 또 다른 세트를 형성한다. 이와 같은 의도적인 제품 배치에 시간과 자원을 투자하면 고객의 쇼핑 시간을 줄일 뿐만 아니라, 고객이 구매에 관심이 있는 관련 품목이 무엇인지 상기시켜주기 때문에, 그 과정에서 상점들이 교차 판매를 할 수 있도록 도와준다.

규칙은 개인의 선호도를 추출하지 않고, 오히려 모든 구별되는 거래의 요소들 사이에서 관계를 찾는다. 협업 필터링과 다른 점이 바로 이것이다.
- 협업필터링은 개인의 선호도를 기준으로 하고, 연관규칙은 그 아이템 요소들 자체를 기준으로 바라보는 관점인 듯함. 어렵네..

---

선행과 결과물로 구성되는데, 둘 다 항목 목록이다. 여기서의 함축은 인과관계가 아닌 공동 발생이라는 점에 유의해야한다. 주어진 규칙에서 항목 집합은 선행 항목과 결과 항목의 목록이다.

---
이 두 가지 사이의 연관성의 강도를 이해하는 데 도움이 되는 다양한 지표가 마련되어 있다.
- Support
- Confidence
- Lift

## Support(지지도)
모든 거래에서 항목 집합이 얼마나 자주 발생하는지를 알 수 있다. [빵,버터] > [빵,샹품]처럼, 수학적으로 Support는 항목 집합이 발생하는 총 트랜잭션 수의 일부분이다.

1_bqdq-z4Ec7Uac3TT3H_1Gg.png

Support는 추가 분석을 위해 고려해야할 가치가 있는 규칙을 식별하는 것에 도움이 많이 됨. 예를 들어, 총 10,000개의 거래 중 최소한 50배 이상 발생하는 항목 집합(예: Support = 0.005)만 고려하고자 할 수 있다.

## Confidence(신뢰도)
카트가 이미 선행 조건을 가지고 있다는 점을 감안하여 카트에서 발생하는 결과의 발생에 대한 호감도를 정의. 즉, {피자}가 포함된 모든 거래 중 {콜라}도 포함된 거래는 몇 개인가? 우리는 상식적으로 {피자} → {콜라}가 높은 신뢰 규칙이 되어야 한다고 말할 수 있다. 기술적으로 신뢰도는 선행조건이 주어진 결과 발생의 조건부 확률이다.



1_E3mNKHcudWzHySGMvo_vPg.png

{Butter} -> {Bread}의 신뢰도는 얼마나 될까? 즉, 버터를 사용하는 거래 중 빵이 포함된 것은 무엇인가? 매우 높음, 즉 1에 가까운 값일 것이다. {요거트} → {우유}는? 높을 것이다. {칫솔} → {우유}? {우유}는 매우 빈번한 항목 집합이며 다른 모든 트랜잭션에 존재하므로 이 규칙에 대한 신뢰도도 높을 것이다.


결국, **빈번한 결과에 대해 선행조건에 무엇을 가지고 있는지는 중요하지 않다는 것을 알아야함.** 연결 규칙의 결과 발생 빈도가 매우 높은 신뢰도는 항상 신뢰도가 높을 것이다.

1_cJqlXK7RrdOd6OhvEsXevw.png

{칫솔} → {우유}
- 10/(10+4) = 0.7

신뢰도가 높은 것 같다. 그러나 직감적으로 이 두 제품의 연관성이 약하고 이 높은 신뢰도가 뭔가 문제점이 있을 것 같다. 여기서 Lift(향상도)는 이 도전을 극복하기 위해 도입된다.

## Lift(향상도)
향상도는 주어진 {X}의 조건부 {Y}를 계산하는 동안 결과의 지원(빈도)를 제어한다.

lift.png

{X}가 {Y}로 카트로 실제로 이어지는 경우, 리프트 값은 1보다 커진다.

- 칫솔이 있을 때, 우유가 있을 신뢰도 :10/14 = 0.7이다.
- 우유가 장바구니에 들어갈 확률: 80/100 = 0.8

카트에 칫솔을 가지고 있는 것이 실제로 카트에 우유를 가지고 있을 확률을 0.8에서 0.7로 줄인다는 것을 보여준다. 이것은 0.7/0.8 = 0.87로 바뀔 것이다.
향상도 값이 1 미만이기 때문에, 칫솔과 우유가 음의 관계인 것을 판단할 수 있음. 향상도가 1 이상이면, 높은 연관성이 있다. 향상도는 물건배치를 결정하는데 큰 도움이 된다.

--- 
상품이 많아지면 지지도,신뢰도,향상도를 계산량이 많아지게 되고 비효율적으로 될 것이다.
