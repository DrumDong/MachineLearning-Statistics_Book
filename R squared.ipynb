## R-squared(결정계수)는 비선형 회귀분석에 유효하지 않다..!

비선형 회귀 분석은 데이터에 존재하는 대부분의 곡선을 적합시킬 수 있는 매우 유연한 분석이다. 결정계수는 회귀 모형의 적합도를 평가하는 매우 직관적인 방법으로 보인다. 하지만 비선형 회귀와 결정계수는 함께할 수 없다. 결정계수는 비선형 회귀 분석에 적합하지 않다.

> 일부 통계 소프트웨어는 통계적으로 정확하지 않더라도 이러한 모델에 대해 결정계수를 계산한다. 따라서 선형 모형이 아닌 모형에 대해 결정계수를 신뢰하지 않는 이유를 이해하는 것이 중요하다. 이번 포스팅에서는 비선형 회귀 분석을 위한 결정계수를 평가하는 것이 심각한 문제를 야기하고, 잘못된 길로 인도하는 방법이라는 것을 알려줄 것이다.



### 왜 결정계수는 오직 선형회귀에서만 타당한 것일까?
R2공식.png

> Explained variance + Error variance = Total variance

이 배열은 항상 0~100% 사이의 결정계수 생성한다.

선형 모델일 경우, 이러한 방식은 예상대로 잘 적용이 된다.

하지만, 이는 오직 선형 회귀 모델에서만 정확하게 작동한다. 비선형 회귀 분석에서는 이러한 근본적인 가정이 부정확하다. 설명된 오차 + 설명되지 않는 오차를 총 분산에 합산하면 안된다. 결과는 결정계수가 반드시 0과 100% 사이는 아니라는 것이다. 여기에는 다른 문제들도 있다.

이 문제는 비선형 회귀 분석의 맥락에서 결정계수를 완전히 약화시킨다. 결정계수는 다항식을 사용하여 곡면성을 모형화하는 선형 모델에 유효하다.
