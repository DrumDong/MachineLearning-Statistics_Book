# Small dataset을 다루는 7개의 효과적인 방법들

사용 가능한 데이터가 부족하더라도 과대 적합을 방지하고 정확한 예측을 얻는 방법에 대해 알아보자.

## Bias-Variance Tradeoff
데이터를 설명하는 완벽한 모델을 찾는 방법은 편향(Bias)과 분산(Variance)의 균형을 찾는 것이다. 

편향은 실제값과 예측값의 차이이다. 편향이 높은 모형은 예측 변수와 목표 변수 간의 관계를 지나치게 단순화하고 학습 및 테스트 데이터에서 모두 높은 오류를 나타낸다. 

분산은 모델 예측의 변동성을 나타낸다. 높은 분산의 모델은 학습 데이터에 너무 치중이 되어있어 테스트 데이터에 일반화를 잘 못한다. 그 결과, 이러한 모델들은 학습 데이터에는 낮은 에러를, 테스트 데이터에서는 높은 에러를 보여준다.

**Overfitting vs. Underfitting vs. Good Balance (source)**

낮은 편향과 높은 분산의 모델은 데이터에 과적합(Overfitting)을 하는 반면에, 높은 편향과 낮은 분산의 모델은 데이터에 과소적합(Underfitting)한다.




작은 데이터셋에서 훈련된 모델은 존재하지 않는 패턴을 더 잘 볼 수 있으며, 이는 테스트셋에 높은 분산과 매우 높은 오차를 초래한다. 이는 과대적합의 일반적인 신호이다. 따라서 소규모 데이터셋으로 작업할 때 가장 중요한 목표는 과대적합을 피하는 것이다.

## 작은 데이터셋에서 과대적합을 해결하는 기술들
이제 소규모 데이터셋으로 작업할 때 과대적합을 방지할 수 있는 가장 유용한 7가지 기술에 대해 알아보자.

#### 1. 간단한 모델을 선택한다.
많은 파라미터가 있는 모델들은 과대적합이 되기 쉽다.
- 분류문제이면 로지스틱 회귀부터!
- 특정 값을 예측하도록 모델을 훈련시키는 경우, 가중치 수가 제한된 단순 선형 모델을 고려하자.
- Tree 모델은, Maxium depth를 정해주자.
- 정규화 기법을 적용하자.

부족한 데이터로, 우리의 목표는 존재하지 않는 패턴과 관계를 볼 수 있는 모델의 능력을 제한하는 것이다.
즉, 가중치와 파라미터의 수를 제한하고 비선형성 또는 변수 교호작용을 암시하는 모든 모델을 배제하려는 것이다.

#### 2. 아웃라이어를 제거하자.
작은 데이터를 다룰 때는, 이상치(Outlier)가 모델에 큰 영향을 끼친다. 그래서, 작은 데이터를 다룰 때는 이상치를 제거해야한다. 또 다른 접근방식은 퀀텀 회귀 분석과 같이 특이치에 강한 기법을 사용하는 것일 수 있다. 특이치의 영향을 제거하는 것은 작은 데이터 세트로 민감한(Sensible) 모델을 얻는 데 필수적이다.

### 3. 변수 선택 적용
일반적으로 명시적인 변수를 선택하는 것이 최선의 방법은 아니지만, 데이터가 제한된 경우 이는 필수적인 단계가 될 수 있다. 소수의 관측치(row)와 많은 수의 예측 변수들(columns)로 과대적합을 피하기는 어렵다. 변수 선택에는 종속 변수와의 상관관계 분석, 중요도 분석 및 반복 제거 등 여러 가지 접근법이 있다. 변수 선택은 항상 도메인 전문지식의 혜택을 받을 것이라는 점도 주목할 필요가 있다. 주제 영역에 익숙하지 않은 경우, 도메인 전문가를 찾아 기능 선택 프로세스에 대해 논도록 하자.

#### 4. 여러 모델을 결합하자.
두 개 이상의 모델에서 결과를 결합하면 훨씬 더 정확한 예측을 얻을 수 있다. 다양한 개별 모델의 예측에 대한 가중 평균으로 계산된 최종 예측은 각 개별 모델의 예측에 비해 분산이 유의하게 낮고 일반성이 개선될 것이다. 또한 서로 다른 하이퍼 파라미터 값을 사용하여 서로 다른 모델 또는 동일한 모델의 예측을 결합할 수 있다.

#### 5. 점 추정 대신 신뢰 구간에 의존해라.
예측 자체 외에도 예측에 대한 신뢰 구간을 추정하는 것이 종종 써먹을 만한 좋은 방법이기도 하다. 이 방법은 특히 소규모 데이터셋으로 작업할 때 중요해진다. 따라서 회귀 분석에서 95% 신뢰 구간을 추정하면 된다. 분류 문제를 해결하는 경우 클래스 예측의 확률을 계산하면 된다. 모델이 예측에 대해 얼마나 "신뢰성"을 갖고 있는지를 더 잘 이해하게 되면, 모형의 결과에 근거한 잘못된 결론에 도달할 가능성이 줄어든다.

#### 6. 데이터를 늘리자.
데이터가 실제로 부족하거나 데이터 집합의 균형이 크게 맞지 않을 경우, 데이터 집합을 확장하는 방법이 있다.
- 합성 샘플을 사용한다.
   - 이것은 데이터 집합에서 특정 클래스의 과소표시를 다루기 위한 일반적인 접근방식이다. 합성 샘플로 데이터 집합을 증가시키는 몇 가지 접근법이 있다. 특정 Task에 가장 적합한 것을 선택하면 된다.
- 데이터 수집
 - 예를 들어 특정 지역의 온도를 모델링하는 경우 다른 지역의 날씨 정보도 함께 사용하십시오. 그러나 관심 영역의 데이터 지점에 더 높은 가중치를 할당하면 된다.

#### 7. 가능하다면 전이학습(transfer learning)을 적용해라.
이 접근법도 데이터 확장의 한 유형이다. 전이 학습은 사용 가능한 대규모 데이터셋에 대한 범용 모델을 교육한 다음 소규모 데이터셋에서 세부적으로 조정하는 것을 의미한다. 예를 들어, 이미지 분류 문제를 작업하는 경우 ImageNet에서 미리 학습된 모델을 사용하여 특정 문제에 맞게 미세 조정을 하면 된다. 사전 훈련된 모델은 제한된 데이터를 사용하여 처음부터 훈련된 모델보다 정확한 예측을 산출할 가능성이 더 높다. 전이학습은 특히 유연한 딥러닝 기법으로 효과가 좋다.

# 작은 데이터는 우리가 극복할 수 있는 다른 Challenge일 뿐이다.
많은 연구자들과 실무자들은 작은 데이터가 데이터 과학의 미래라고 믿는다. 모든 종류의 문제에 대해 거대한 데이터 세트를 갖는 것은 실현 가능하지 않다. 소규모 데이터셋의 문제를 해결하려면 다음 지침을 따르도록 하자.
- 적은 관측치를 처리할 때 어떤 종류의 문제를 예상할 수 있는지 알기 위해 통계의 기본 사항을 이해하자.
- 너무 많은 데이터를 포함하지 않고 작은 데이터에서 정확한 결과를 얻을 수 있는 주요 전략을 알아보자.
- 모든 데이터 처리 및 분석 단계를 효율적으로 수행
- 예측에서 결론을 도출할 때 모델의 한계에 유의하자.
